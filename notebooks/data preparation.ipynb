```python
import pandas as pd
import librosa
import os
from tqdm import tqdm
import numpy as np
import multiprocessing
import glob
tqdm.pandas()
```

```python
data_folder = "../data"
```

```python
df21 = pd.read_csv(os.path.join(data_folder, "birdclef-2021/train_metadata.csv"))
df21["filename"] = df21["primary_label"] + "/" + df21["filename"]
df21["path"] = os.path.join(data_folder, "birdclef-2021/train_short_audio") + "/" + "/" + df21["filename"]
df21["path_exist"] = df21["path"].apply(os.path.exists)
df21.head()
```

```python
df22 = pd.read_csv(os.path.join(data_folder, "birdclef-2022/train_metadata.csv"))
df22["path"] = os.path.join(data_folder, "birdclef-2022/train_audio") + "/" + df22["filename"]
df22["path_exist"] = df22["path"].apply(os.path.exists)
df22.head()
```

```python
df23 = pd.read_csv(os.path.join(data_folder, "birdclef-2023/train_metadata.csv"))
df23["path"] = os.path.join(data_folder, "birdclef-2023/train_audio") + "/" + df23["filename"]
df23["path_exist"] = df23["path"].apply(os.path.exists)
df23.head()
```

```python
df24 = pd.read_csv(os.path.join(data_folder, "birdclef-2024/train_metadata.csv"))
df24["path"] = os.path.join(data_folder, "birdclef-2024/train_audio") + "/" + df24["filename"]
df24["path_exist"] = df24["path"].apply(os.path.exists)
df24.head()
```

```python
merged_df = pd.concat([df21,df22,df23,df24]).reset_index(drop=True)
merged_df.shape
```

```python
def get_duration(filename):
    y, sr = librosa.load(filename, sr=32000, mono=True)
    return librosa.get_duration(y=y, sr=sr)
```

```python
merged_df["duration"] = merged_df["path"].progress_apply(get_duration)
```

```python
merged_df.head()
```

```python
merged_df.to_csv(os.path.join(data_folder, "4_comps_training_metdata.csv"), index=False)
```
